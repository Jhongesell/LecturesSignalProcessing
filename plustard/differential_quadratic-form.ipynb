

On cherche à dériver le critère quadratique suivant
\begin{equation}
C(\theta)=\left(y-H\theta\right)^{T}M\left(y-H\theta\right)
\end{equation}
par rapport au vecteur colonne $\theta$ de dimension $N$. 

Débutons par l'examen de la forme quadratique simple $z^{T}Mz$. En
développant le produit, on a $z^{T}Mz=\sum_{i=1}^{N}\sum_{j=1}^{N}M_{ij}z_{i}z_{j}$.
Par suite la dérivée par rapport à la $k^{e}$composante $z_{k}$s'écrit
:
\begin{equation}
\frac{{\ensuremath{\partial}}}{\partial z_{k}}\sum_{i=1}^{N}\sum_{j=1}^{N}M_{ij}z_{i}z_{j}=\sum_{i=1}^{N}M_{ik}z_{i}+\sum_{j=1}^{N}M_{kj}z_{j}=\left[M^{T}z\right]_{k}+\left[Mz\right]_{k},
\end{equation}
où $[]_{k}$désigne la $k^{e}$composante du vecteur en question.
La dérivée par raport à un vecteur est tout simplement le vecteur
des dérivées par rapport à chacune des composantes. On le note comme
l'application de l'opérateur 
\begin{equation}
{{\frac{\text{d}}{\text{d}x}}={\left[\begin{array}{c}
 \frac{{{\partial}}}{\partial x_{1}}\\
\frac{{{\partial}}}{\partial x_{2}}\\
\vdots\\
\frac{{{\partial}}}{\partial x_{N}}
\end{array}\right]}}.
\end{equation}
Du coup, on aura
\begin{equation}
\frac{\text{d}}{\text{d}z}z^{T}Mz=(M+M^{T})z
\end{equation}
et si $M$ est symétrique, $M=M^{T},$ alors $\frac{\text{d}}{\text{d}z}z^{T}Mz=2Mz$.
Bien entendu on retrouve les résultats bien connus dans le cas scalaire.
On notera que la dérivée d'un scalaire par rapport à un vecteur fournit
un vecteur et que la dérivée d'un vecteur ligne par rapport à un vecteur
conduit de son côté à une matrice. 

Supposons maintenant que $z$ dépende de $x$ et que l'on effectue
le changement de variable $z\rightarrow x.$ il est bien connu que
$$
\frac{{{\partial}}}{\partial x_{1}}=\frac{\partial z_{1}}{\partial x_{1}}\frac{\partial}{\partial z_{1}}+\ldots+\frac{\partial z_{N}}{\partial x_{1}}\frac{\partial}{\partial z_{N}},
$$
 où $\left[\frac{\partial z_{1}}{\partial x_{1}}\ldots\frac{\partial z_{N}}{\partial x_{1}}\right]$
est la forme Jacobienne du changement de variable. Dans ces conditions,
on voit que 
\begin{equation}
{{\frac{\text{d}}{\text{d}x}}={\left[\begin{array}{c}
 \frac{{{\partial}}}{\partial x_{1}}\\
\frac{{{\partial}}}{\partial x_{2}}\\
\vdots\\
\frac{{{\partial}}}{\partial x_{N}}
\end{array}\right]}}=\left[\begin{array}{cccc}
\frac{\partial z_{1}}{\partial x_{1}} &  &  & \frac{\partial z_{N}}{\partial x_{1}}\\
\frac{\partial z_{1}}{\partial x_{2}} &  &  & \frac{\partial z_{N}}{\partial x_{2}}\\
\\
\frac{\partial z_{1}}{\partial x_{N}} &  &  & \frac{\partial z_{N}}{\partial x_{N}}
\end{array}\right]{{\left[\begin{array}{c}
 \frac{{{\partial}}}{\partial z_{1}}\\
\frac{{{\partial}}}{\partial z_{2}}\\
\vdots\\
\frac{{{\partial}}}{\partial z_{N}}
\end{array}\right]}}=J\frac{\text{d}}{\text{d}z},
\end{equation}
où $J,$ le Jacobien, est simplement $J=\frac{dz^{T}}{dx}$. 

Revenons à notre crière quadratique initial. On a $z=y-H\theta$,
et $\frac{dz^{T}}{d\theta}=\frac{d}{d\theta}\left(y^{T}-\theta^{T}H^{T}\right)=0-H^{T}$
(car vous vérifierez aisément que $\frac{dz^{T}}{d\theta}=1$, la
matrice identité). Par suite, 
\begin{equation}
\frac{d}{d\theta}\left(y-H\theta\right)^{T}M\left(y-H\theta\right)=-H^{T}(M+M^{T})\left(y-H\theta\right)
\end{equation}
ou encore 
\begin{equation}
\frac{d}{d\theta}\left(y-H\theta\right)^{T}M\left(y-H\theta\right)=-2H^{T}M\left(y-H\theta\right)\text{ si $M$ est symétrique. }
\end{equation}


Allez, tant qu'à faire : 

La valeur de $\theta$ qui minimise le critère quadratique est telle
que
\begin{equation}
-2H^{T}M\left(y-H\theta\right)=0,
\end{equation}
soit en développant : $H^{T}My=H^{T}MH\theta$. Si la matrice $(H^{T}MH)$
a le bon goût d'être inversible, on aura alors 
\begin{equation}
\theta=(H^{T}MH)^{-1}H^{T}My
\end{equation}
qui est la solution des moindres carrés pondérés (la matrice $M$
est la matrice de pondération).
